```python
#starting_animal_island

# Using fruithful function(Returns a calculated relationship)
def calculated_bond(level, bonus):
    if level >= 5:
        return level + bonus
    else: 
        return level

# A void function was used to print a news
def show_island_news(text):
    print("ISLAND NEWS: " + str(text))

print("--- Welcome to Animalia! ----")

# Get to know islander profiles 
islander1_name = "Missy Cat"
islander2_name = "Rolf Dog"
friend_level = int("3") # String converted to Integer
happiness_percentage = float(80.0) # Integer coverted to float

# Get user input for an action
print("What should they do today?")
print("Action: 'Gift' to give them a present, or 'prank' to cause trouble.")
action = input("Enter your choice: ")

# Using two-way decision for the action
if action == "gift":
    print("You gave them a toy! They played together all day.")
    friend_level = friend_level + 3
else:
    print("Oh no! A thief was spotted! Someone stole the other's favorite snack!")
    friend_level = friend_level - 2

# Calling fruitful function with arguments
final_bond = calculated_bond(friend_level, 2)

# Checking the outcome 
if final_bond >= 5: 
    happiness_percentage = happiness_percentage + 15.5
    show_island_news(islander1_name + " and " + islander2_name + " became Best friends! Nice one!")
else:
     happiness_percentage = happiness_percentage - 20.0                     
     show_island_news("Uh ho ... Someone needs an apology.")

# Final checking
print("Island Happiness level: " + str(happiness_percentage))
```
